## Question 1
Print numbers from 1 to 50. Skip multiples of 3 and stop if a number divisible by both 5 and 7 appears.

```python
for i in range(1, 51):
    if i % 5 == 0 and i % 7 == 0:
        break
    if i % 3 == 0:
        continue
    print(i, end=' ')
```

## Question 2
Take a number and count frequency of each digit without converting it to string.

```python
num = int(input("Enter a number: "))
temp = abs(num)
freq = {}

while temp > 0:
    digit = temp % 10
    freq[digit] = freq.get(digit, 0) + 1
    temp //= 10

print("Digit frequency:")
for digit, count in sorted(freq.items()):
    print(f"{digit}: {count}")
```

## Question 3
Take a number and reverse it, but stop reversing if digit 0 appears.

```python
num = int(input("Enter a number: "))
temp = abs(num)
reversed_num = 0

while temp > 0:
    digit = temp % 10
    if digit == 0:
        break
    reversed_num = reversed_num * 10 + digit
    temp //= 10

print(f"Reversed number: {reversed_num}")
```

## Question 4
Print pattern with numbers in increasing order but skip numbers divisible by 4.

```python

rows = int(input("Enter number of rows: "))
num = 1

for i in range(1, rows + 1):
    for j in range(i):
        while num % 4 == 0:
            num += 1
        print(num, end=' ')
        num += 1
    print()
```

## Question 5
Take a number and compare it with its reversed number and classify as Increasing or Decreasing.

```python
num = int(input("Enter a number: "))
temp = abs(num)
reversed_num = 0

while temp > 0:
    digit = temp % 10
    reversed_num = reversed_num * 10 + digit
    temp //= 10

if num > reversed_num:
    print("Increasing")
elif num < reversed_num:
    print("Decreasing")
else:
    print("Same")
```

## Question 6
Print numbers from 1–30:
- Even → square
- Odd → sum from 1 to that number

```python
for i in range(1, 31):
    if i % 2 == 0:
        print(f"{i}: {i**2}")
    else:
        sum_val = sum(range(1, i + 1))
        print(f"{i}: {sum_val}")
```

## Question 7
Find the first non-repeating digit in a number.

```python
num = input("Enter a number: ")
found = False

for digit in num:
    if num.count(digit) == 1:
        print(f"First non-repeating digit: {digit}")
        found = True
        break

if not found:
    print("No non-repeating digit found")
```

## Question 9
Take inputs until 0 is entered:
- Add even numbers
- Subtract odd numbers

```python
result = 0

while True:
    num = int(input("Enter a number (0 to stop): "))
    if num == 0:
        break
    if num % 2 == 0:
        result += num
    else:
        result -= num

print(f"Final result: {result}")
```

## Question 10
Print star pattern but skip star at position (row=4, col=2).

```python
rows = int(input("Enter number of rows: "))

for i in range(1, rows + 1):
    for j in range(1, i + 1):
        if i == 4 and j == 2:
            print(' ', end=' ')
        else:
            print('*', end=' ')
    print()
```

## Question 11
Find Armstrong numbers from 1–500 with early loop exit optimization.

```python
print("Armstrong numbers from 1 to 500:")

for num in range(1, 501):
    temp = num
    sum_val = 0
    digits = len(str(num))

    while temp > 0:
        digit = temp % 10
        sum_val += digit ** digits
        if sum_val > num:  # Early exit optimization
            break
        temp //= 10

    if sum_val == num:
        print(num, end=' ')
```

## Question 12
Count consecutive duplicates in a list.

```python
lst = [1, 1, 2, 2, 2, 3, 4, 4, 5, 5, 5, 5]
count = 0

for i in range(len(lst) - 1):
    if lst[i] == lst[i + 1]:
        count += 1

print(f"Number of consecutive duplicates: {count}")
```

## Question 13
Create a password system with 3 attempts and conditional hints.

```python
password = "python123"
attempts = 3

for i in range(attempts):
    user_input = input(f"Enter password (Attempt {i + 1}/{attempts}): ")

    if user_input == password:
        print("Access granted!")
        break
    else:
        remaining = attempts - i - 1
        if remaining > 0:
            print(f"Incorrect password. {remaining} attempt(s) remaining.")
            if i == 0:
                print("Hint: Password contains 'python'")
            elif i == 1:
                print("Hint: Password has numbers too")
        else:
            print("Access denied! No more attempts.")
```

## Question 14
Check if a number is prime with early loop break.

```python
num = int(input("Enter a number: "))

if num < 2:
    print(f"{num} is not prime")
else:
    is_prime = True
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            is_prime = False
            break

    if is_prime:
        print(f"{num} is prime")
    else:
        print(f"{num} is not prime")
```

## Question 15
Print number pattern replacing multiples of 3 with *.

```python
rows = int(input("Enter number of rows: "))
num = 1

for i in range(1, rows + 1):
    for j in range(i):
        if num % 3 == 0:
            print('*', end=' ')
        else:
            print(num, end=' ')
        num += 1
    print()
```
