# Python Assignment: If-Else Conditionals



## Question 1
Create a program that compares two numbers and prints the larger one using if-else.

```python
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

if num1 > num2:
    print(f"The larger number is: {num1}")
elif num2 > num1:
    print(f"The larger number is: {num2}")
else:
    print("Both numbers are equal")
```

## Question 2
Create a program that classifies a given temperature as:
- Hot (>30°C)
- Warm (20-30°C)
- Cold (<20°C)

Using if-elif-else statements.

```python
temperature = float(input("Enter temperature in Celsius: "))

if temperature > 30:
    print("Hot")
elif temperature >= 20 and temperature <= 30:
    print("Warm")
else:
    print("Cold")
```

## Question 3
Create a Python program that checks if a given string is a palindrome using if-else statement.
Take the input from the user.

```python
string = input("Enter a string: ")

# Remove spaces and convert to lowercase for better palindrome checking
cleaned_string = string.replace(" ", "").lower()
reversed_string = cleaned_string[::-1]

if cleaned_string == reversed_string:
    print(f"'{string}' is a palindrome")
else:
    print(f"'{string}' is not a palindrome")
```

## Question 4
Create a Python program that takes a letter as input and determines if it's a vowel or a consonant using if-else.

```python
letter = input("Enter a letter: ").lower()

if len(letter) == 1 and letter.isalpha():
    if letter in 'aeiou':
        print(f"'{letter}' is a vowel")
    else:
        print(f"'{letter}' is a consonant")
else:
    print("Please enter a single letter")
```

## Question 5
Create a Python program that checks if the email ID is valid or not.
If your email contains @ and . then it is valid.
Take the email data from user.

```python
email = input("Enter the email: ")

if "@" in email and "." in email:
    print("Valid email")
else:
    print("Invalid email")
```
